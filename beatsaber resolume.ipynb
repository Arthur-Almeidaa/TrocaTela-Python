```python
import os
import cv2
import numpy as np
from PIL import Image
from time import sleep
from tqdm import tqdm
import pyautogui
import requests
from keras.models import load_model
import schedule
import matplotlib.pyplot as plt

os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'
SIZE = 150

print('Iniciando o modelo...')


model = load_model('model150px.h5')

model.compile(loss='binary_crossentropy',
             optimizer='Adam',
             metrics=['accuracy'])

status = []

def screenshot():
    # processamento de imagens
    
    pyautogui.screenshot('teste.png')     
    image = cv2.imread('teste.png')
    image = Image.fromarray(image, 'RGB')
    image = image.resize((SIZE, SIZE))
    input_image = np.expand_dims(image, axis=0)
    
    # predict
    sleep(1)
    predict = model.predict(input_image)
    print(predict)
    
    
    # envio de informações
    if predict == 1:
        print('Jogando')
        status.append(1)
    else:
        print('Parado')
        status.append(0)
    
    if len(status) >= 3:
        
        # trocar a tela
        if status[0] == 0 and status[1] == 0:
            layer_index = 2
            clip_index = 1 
            
            base_url = f'http://10.0.0.102:8080/api/v1/composition/layers/{layer_index}/clips/{clip_index}/connect'
            
            response = requests.post(base_url)
            print(response)
        elif status[1] == 0 and status[2] == 0:
            layer_index = 2
            clip_index = 1 
            
            base_url = f'http://10.0.0.102:8080/api/v1/composition/layers/{layer_index}/clips/{clip_index}/connect'
            response = requests.post(base_url)
            print(response)
            
        # manter a tela
        elif status[0] == 1 and status[1] == 1:
            layer_index = 2
            clip_index = 2 
            
            base_url = f'http://10.0.0.102:8080/api/v1/composition/layers/{layer_index}/clips/{clip_index}/connect'
            response = requests.post(base_url)
            print(response)
            
        elif status[1] == 1 and status[2] == 1:
            layer_index = 2
            clip_index = 2 
            
            base_url = f'http://10.0.0.102:8080/api/v1/composition/layers/{layer_index}/clips/{clip_index}/connect'
            response = requests.post(base_url)
            print(response)
            
        status.pop(0)
        
        print(status)

print('Schedule iniciando...')

while True:
    print('Função chamada')
    screenshot()
```
